Patterns In Python

```python
# 1. Start square pattern


size = 5

for i in range(size):
    for j in range(size):
        print("*", end=" ")
    print()
```

```python
# 2. Right star triangle 

def right_star_triangle(n):
    for i in range(1, n + 1):

        print(" " * (n - i) + "*" * i)

rows = 5
right_star_triangle(rows)

```

```python
# 3. Right Rt triangle(space)

n = 5
for i in range(1, n + 1):
    print(" " * (n - i) + "*" * i)
```

```python
# 4. Number pattern (Constant)

rows = 5
num = 3

for i in range(rows):
    for j in range(rows):
        print(num, end=" ")
    print()

```

```python
# 5. Increasing Number pattern

rows = 5
for i in range(1, rows + 1):
    for j in range(i):
        print(i, end=" ")
    print()
```

```python
# 6. Alphabet pattern 

rows = 5
ascii_value = 65  # ASCII for 'A'

for i in range(rows):
    for j in range(i + 1):
        print(chr(ascii_value), end=" ")
    ascii_value += 1
    print()
```

```python
# 7. Merging of the pattern

list_a = [1, 3, 5]
list_b = [2, 4, 6]
merged_list = [item for pair in zip(list_a, list_b) for item in pair]

print(f"Interleaved Pattern: {merged_list}")
```

```python
# 8. Advanced pattern

def print_spiral(n):
    size = 2 * n - 1
    for i in range(size):
        for j in range(size):
            dist = min(i, j, size - i - 1, size - j - 1)
            print(n - dist, end=" ")
        print()

print_spiral(4)
```
