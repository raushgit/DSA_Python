```python
#Demonstrate the type() to show the various data types.

num_int = 10
num_float = 10.5

str1 = "Raushan"

is_yes = True

my_list = [1, 2, 3]        
my_tuple = (1, 2, 3)    

my_dict = {"name": "Raushan", "age": 21}

my_set = {1, 2, 3}


# --- Demonstrating type() ---
print(num_int, type(num_int))
print(num_float, type(num_float))
print(str1, type(str1))
print(is_yes, type(is_yes))
print(my_list, type(my_list))
print(my_tuple, type(my_tuple))
print(my_dict, type(my_dict))
print(my_set, type(my_set))
```
