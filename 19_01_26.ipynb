```python
# WAP to check if the input number is Armstrong or not.

num = int(input("Enter a number: "))

num_str = str(num)
power = len(num_str)

total_sum = 0
for digit in num_str:
    total_sum += int(digit) ** power
    
if total_sum == num:
    print(num ,"is an Armstrong number.")
else:
    print(num ,"is NOT an Armstrong number.")
```

```python
# WAP to print even numbers in a given sequence(using list) using a for loop.

my_numbers = [12, 7, 34, 23, 89, 4, 90, 15]

print("Even numbers in the list are:")

for num in my_numbers:
    if num % 2 == 0:
        print(num)
```

```python
# Print the multiple of 10, in a given range entered by the user. (Note: Do this question using range()).

start = int(input("Enter start of range: "))
end = int(input("Enter end of range: "))

print(f"Multiples of 10 between start and end")

for num in range(start, end + 1):
    if num % 10 == 0:
        print(num)

```

Which data type will be used to represent the following data values and why?


1. Number of months in a year -> int
Reason - This is a countable, whole number (always 12) that cannot have a decimal point.

2. Resident of Delhi or not -> bool
Reason - This represents a binary condition (True/False or Yes/No).

3. Mobile number -> str
Reason - While it looks like a number, it is not used for calculations. Storing it as a string preserves leading zeros and allows country codes.

4. Pocket money -> float
Reason  - Currency amounts often contain decimal parts. Even if it is a whole number, money is best treated as a float for accuracy.

5. Volume of a sphere -> float 
Reason - The formula involves $\pi$ ($3.14159...$) and division, which almost always results in a number with decimals.

6. Perimeter of a square -> float or int
Reason - Measurements are generally treated as float to allow for precision.

7. Name of the student -> str
Reason - Names are sequences of alphanumeric characters.

8. Address of the student -> str
Reason - Addresses contain a mix of letters, numbers, spaces, and special characters.
